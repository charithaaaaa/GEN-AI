```python
# ===== 5. RAG QUERY =====
def search_meeting_notes(query: str, n_results: int = 2):
    results = collection.query(
        query_texts=[query],
        n_results=n_results
    )
    for doc, meta in zip(results['documents'][0], results['metadatas'][0]):
        print(f"\n📄 Meeting ({meta['type']}):\n{doc}\n{'━'*50}")

# Example usage:
search_meeting_notes("What was decided about AI ethics?")
```
